# Ridge-Lasso regression

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler().fit(X_train)
#Standardizing the data set and creating scaler
#scaler is storing the info of how to divide each row and each cell
```

```python
X_train_s = scaler.transform(X_train)
#transforming X_train to scaler trainset
```

```python
X_test_s = scaler.transform(X_test)
#transforming X_test to scaler testset
```

```python
from sklearn.linear_model import Ridge
```

```python
lm_r = Ridge(alpha = 0.5)#here lambda is alpha = 0.5
lm_r.fit(X_train_s, y_train)
```

```python
r2_score(y_test, lm_r.predict(X_test_s))
```

```python
from sklearn.model_selection import validation_curve
```

```python
param_range = np.logspace(-2,8,100)
#array for the lambda values
#this logspace() will create 100 values btw 10 rasied to 8 and 10 rasied to -2
```
